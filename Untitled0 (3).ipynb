We are going to solve 1d schrodinger equation. This Python program solves the one-dimensional time-independent Schr√∂dinger equation for a particle in a hyperbolic potential well. It uses the Numerov method, a powerful numerical tool for solving differential equations, to compute the wavefunctions, and employs root-finding techniques to determine the energy eigenvalues. # **Constant Parameters**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root_scalar

# Constants and Parameters
hbar, m, alpha, lmbda = 1, 1, 1, 4  # hbar=1, mass=1, alpha=1, lambda=4
```

Here,‚Ñè (Planck‚Äôs constant) and m (particle mass) are set to 1 to simplify the equations. The parameters ùõº and Œª define the scale and depth of the potential. For this example,ùõº=1 determines the width of the potential well, and ùúÜ=4 adjusts its depth.


# **Defining Potential Function**


```python
# Potential Function
def V(x):
    """Defines the potential V(x)"""
    coeff = (hbar**2 * alpha**2 * lmbda * (lmbda - 1)) / (2 * m)
    return coeff * (0.5 - 1 / (np.cosh(alpha * x) ** 2))
```

# **Numerov Method for Solving the Schr√∂dinger Equation**

The wavefunction

 is computed using the Numerov method, implemented in the **numerov_step** function:

```python
# Numerov Method for Wavefunction
def numerov_step(psi0, psi1, energy, x_values, h):
    """Numerov method to compute the wavefunction."""
    n_points = len(x_values)
    psi = np.zeros(n_points)
    psi[0], psi[1] = psi0, psi1
    f = lambda x_i: 2 * m / hbar**2 * (energy - V(x_i))

    for idx in range(1, n_points - 1):
        k0, k1, k2 = f(x_values[idx - 1]), f(x_values[idx]), f(x_values[idx + 1])
        psi[idx + 1] = ((2 * (1 - (5 / 12) * h**2 * k1) * psi[idx] -
                        (1 + (1 / 12) * h**2 * k0) * psi[idx - 1]) /
                        (1 + (1 / 12) * h**2 * k2))
    return psi
```

# **Wavefunction Matching Condition**

For a valid solution, the wavefunction must be continuous and differentiable throughout the domain. This is ensured by checking the derivatives of the wavefunction at the midpoint of the potential:

```python
# Wavefunction Matching Condition
def match_condition(energy, x_values, h):
    """Matches wavefunction derivatives at the midpoint."""
    psi_left = numerov_step(0.0, 1e-5, energy, x_values, h)
    psi_right = numerov_step(0.0, 1e-5, energy, x_values[::-1], h)[::-1]
    mid_idx = len(x_values) // 2
    left_deriv = (psi_left[mid_idx + 1] - psi_left[mid_idx - 1]) / (2 * h * psi_left[mid_idx])
    right_deriv = (psi_right[mid_idx + 1] - psi_right[mid_idx - 1]) / (2 * h * psi_right[mid_idx])
    return left_deriv - right_deriv
```

The function integrates the wavefunction from both ends of the domain and computes the derivatives at the midpoint. The condition is satisfied if the derivatives from the left and right match:

# **Finding Eigenvalues**

The eigenvalues are determined using a root-finding technique:

```python
# Find Eigenvalues
def compute_eigenvalues(x_values, h, levels=3):
    """Finds the eigenvalues using the root-finding method."""
    eigenvals, energy_ranges = [], [(-5, 0), (0, 5), (2, 5)]
    for low, high in energy_ranges[:levels]:
        energy_guesses = np.linspace(low, high, 100)
        for e1, e2 in zip(energy_guesses[:-1], energy_guesses[1:]):
            if match_condition(e1, x_values, h) * match_condition(e2, x_values, h) < 0:
                result = root_scalar(match_condition, args=(x_values, h), bracket=(e1, e2), method='brentq')
                eigenvals.append(result.root)
                break
    return eigenvals
```

This function searches for roots of the **match_condition** function. The energy domain is split into smaller intervals, and the program checks for sign changes in the matching condition. A sign change indicates that an eigenvalue exists within the interval. The exact root is found using Brent method via **root_scalar**.

# **Preparing the Spatial Grid**

Before solving, the spatial grid and step size are defined:

```python
# Parameters for Integration
x_range = (-10, 10)  # Range for x
h = 0.05
x_grid = np.arange(x_range[0], x_range[1] + h, h)
```

The grid spans from x=‚àí10 to ùë•=10, with small increments h=0.05. A finer grid improves accuracy but increases computation time.

# **Computing and Displaying Results**

The program calculates both numerical and exact eigenvalues:

```python
# Compute Eigenvalues
num_levels = 3
computed_eigenvalues = compute_eigenvalues(x_grid, h, levels=num_levels)
```

The exact eigenvalues are derived from the analytical solution for the hyperbolic potential, where they are available.

Finally, the results are displayed

```python
# Output Results
print("Computed Eigenvalues:")
for idx, eigenval in enumerate(computed_eigenvalues):
    print(f"E_{idx} = {eigenval:.6f}")

# Exact Eigenvalues
exact_eigenvalues = [(hbar**2 * alpha**2 / (2 * m)) * (lmbda * (lmbda - 1) / 2 - (lmbda - 1 - n)**2) for n in range(num_levels)]
print("\nExact Eigenvalues:")
for idx, eigenval in enumerate(exact_eigenvalues):
    print(f"E_{idx} = {eigenval:.6f}")
```

# **Plotting the Potential and Wavefunctions**

The potential and wavefunctions are visualized using matplotlib:


```python
# Plotting Results
plt.figure(figsize=(8, 6))
plt.plot(x_grid, V(x_grid), label="Potential V(x)", color="black")

for idx, energy in enumerate(computed_eigenvalues):
    psi_wave = numerov_step(0.0, 1e-5, energy, x_grid, h)
    plt.plot(x_grid, psi_wave / np.max(np.abs(psi_wave)) + energy, label=f"E_{idx} = {energy:.3f}")

plt.xlabel("x")
plt.ylabel("Energy / Wavefunction")
plt.title("Potential and Wavefunctions")
plt.xlim(-5, 5)
plt.legend()
plt.grid()
plt.show()
```
